# **Training Models**
---


à¹ƒà¸™à¸‚à¸±à¹‰à¸™à¸•à¸­à¸™à¸™à¸µà¹‰à¸ˆà¸°à¹€à¸›à¹‡à¸™à¸à¸²à¸£à¹€à¸—à¸£à¸™à¹‚à¸¡à¹€à¸”à¸¥ à¹‚à¸”à¸¢à¹ƒà¸Šà¹‰à¸‚à¹‰à¸­à¸¡à¸¹à¸¥ à¸—à¸µà¹ˆà¹€à¸£à¸²à¹„à¸”à¹‰à¸„à¸±à¸”à¸à¸£à¸­à¸‡à¸•à¸±à¸”à¹à¸•à¹ˆà¸‡à¹„à¸›à¸„à¸£à¸±à¸š à¹‚à¸”à¸¢à¸ˆà¸°à¹ƒà¸Šà¹‰à¹€à¸›à¹‡à¸™à¹‚à¸¡à¹€à¸”à¸¥ Random Forest ðŸŒ²

```python
from sklearn.preprocessing import StandardScaler

columns_to_scale = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary', 'Satisfaction Score', 'Point Earned', 'WealthnessRatio']

scaler = StandardScaler()

df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

x = df.drop(axis=1,columns=['Exited'])
y = df["Exited"]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

model = RandomForestClassifier(n_estimators=100, random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

# **Model Evalutation**
---

à¹€à¸¡à¸·à¹ˆà¸­à¹€à¸£à¸²à¹„à¸”à¹‰à¹‚à¸¡à¹€à¸”à¸¥à¸—à¸µà¹ˆà¹€à¸£à¸²à¹€à¸—à¸£à¸™à¸¡à¸²à¹à¸¥à¹‰à¸§ à¸à¹‡à¸ˆà¸°à¸—à¸³à¸à¸²à¸£à¸¡à¸²à¸›à¸£à¸°à¹€à¸¡à¸´à¸™à¸œà¸¥à¹‚à¸¡à¹€à¸”à¸¥à¸à¸±à¸™à¸”à¹‰à¸§à¸¢ Confusion Matrix à¹à¸¥à¸° F1-Score à¸„à¸£à¸±à¸š

```python
#à¹à¸ªà¸”à¸‡ Confusion Matrix

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred, labels=model.classes_)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()

plt.show()
```

```python
#à¹à¸ªà¸”à¸‡à¸„à¹ˆà¸² F1-Score

from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred, average='weighted')
f1
```
